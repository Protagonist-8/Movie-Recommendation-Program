```python
import pandas as pd
import re
import json
```

```python
path1="/content/drive/MyDrive/Datasets/tmdb_5000_credits.csv"
path2="/content/drive/MyDrive/Datasets/tmdb_5000_movies.csv"
movies=pd.read_csv(path2)
credits=pd.read_csv(path1)
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['id','title','genres','overview','keywords','cast','crew']]
```

```python
movies.isna().sum()
```

```python
movies=movies.dropna()
```

```python
l=0
for i in movies.genres:
  name_pattern = re.compile(r'"name":\s*"([^"]+)"')
  names = name_pattern.findall(i)
  movies.genres[l]=names
  l=l+1
```

```python
l=0
for i in movies.keywords:
  name_pattern = re.compile(r'"name":\s*"([^"]+)"')
  names = name_pattern.findall(i)
  movies.keywords[l]=names
  l=l+1
```

```python
l=0
for i in movies.cast:
  name_pattern = re.compile(r'"name":\s*"([^"]+)"')
  names = name_pattern.findall(i)
  movies.cast[l]=names[0:3]
  l=l+1
```

```python
l=0
for i in movies.crew:
  list_of_dicts=json.loads(i)
  for j in list_of_dicts:
    if(j["job"]=="Director"):
      movies.crew[l]=[j["name"]]
      break
  l=l+1
```

```python
movies.overview=movies['overview'].apply(lambda x:x.split())
```

```python
#Removing extra spaces so as not to misdirect
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies_new=movies[['id','title','tags']]
```

```python
movies_new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv1=CountVectorizer(max_features=4806,stop_words='english')
```

```python
features_1=cv1.fit_transform(movies_new['tags']).toarray()
```

```python
cv1.get_feature_names_out()
```

```python
cv2=CountVectorizer(max_features=100000,stop_words='english')
```

```python
features_2=cv2.fit_transform(movies_new['tags']).toarray()
```

```python
cv2.get_feature_names_out()
```

```python
cv3=CountVectorizer(max_features=1000,stop_words='english')
```

```python
features_3=cv3.fit_transform(movies_new['tags']).toarray()
```

```python
cv3.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_1=cosine_similarity(features_1)
```

```python
similarity_2=cosine_similarity(features_2)
```

```python
#main function
def recommend(movie):
  movie_index=movies_new[movies_new.title==movie].index[0]
  similar_score=similarity_1[movie_index]
  movies_list=sorted(list(enumerate(similar_score)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(movies_new.iloc[i[0]].title,"Percentage Match:",round(i[1]*100,ndigits=2))
```

```python
#Checking my watched movie is in data or not
'Prisoners' in movies_new.title.values
```

```python
#Asking the model to recommend the similar movies
recommend('Prisoners')
```
